# SEGMENTAÇÃO DE IMAGENS

As técnicas de processamento de imagens começaram a ter visibilidade no período da corrida espacial, em que houve uma grande necessidade de análise e melhoramento das imagens enviadas pelas sondas espaciais, das missões para a Lua. As imagens recebidas pelos satélites e pelas sondas precisavam de tratamento contra distorções e ruídos e foi assim que a área de processamento de imagens ganhou visibilidade e começou a ser aprofundada pelos estudiosos.

O processamento de imagens digitais envolve uma série de conceitos que incluem desde a compressão de imagens, reconhecimento e interpretação de imagens até a segmentação de imagens, assunto que será abordado ao longo desse trabalho e que pode ser dividido em: detecção de bordas, de pontos, de linhas e descontinuidade.

Em suma, segmentação de imagens é uma técnica de análise digital de imagens. Refere-se ao processo de dividir uma imagem em múltiplas regiões (conjunto de pixels uniformes) ou objetos, com o objetivo de simplificar e/ou mudar a sua representação a fim de facilitar sua análise. Como resultado, cada um dos pixels em uma mesma região é similar com referência a alguma característica ou propriedade computacional, tais como cor, intensidade, textura ou continuidade. Regiões adjacentes devem possuir diferenças significativas com respeito a mesma característica(s). A divisão em porções consiste basicamente em um processo de crescimento de regiões, de detecção de bordas ou de detecção de bacias.

Sendo um método tipicamente utilizado para localizar objetos e formas (linhas, curvas, etc) em imagens suas aplicações são diversas. Dentre elas, é possível citar: imagens Médicas; localização de tumores e outras patologias; medida de volume de tecidos; cirurgia guiada por computador; diagnostico de doenças; planos de tratamento; estudo da estrutura anatômica; localização de objetos em imagens de satélite (estradas, florestas, etc.); sistemas de reconhecimento facial; sistemas de controle automático de trafego; sistemas de visão computacional.

Funções utilizadas:

```python
# Função para segmentar utilizando o Kmeans cluster
def segment(image, k):
    
    # Remodelando a imagem para um array 2D de pixels e 3 valores de cores(RGB)
    pixel_vals = image.reshape((-1,3)) 
    
    # Convertendo para float
    pixel_vals = np.float32(pixel_vals)
     
    # A linha a seguir define o critério para o algoritmo parar de rodar,
    # o que acontece quando o algorimto completa 100 iterações ou atinge a precisão de 85% 
    criteria = (cv2.TERM_CRITERIA_EPS + cv2.TERM_CRITERIA_MAX_ITER, 100, 0.85)
    
    # Então o cluestering é realizado de acordo com o número de cluesters definidos por k
    # Centros aleatórios também são escolhidos para o k-means clustering
    retval, labels, centers = cv2.kmeans(pixel_vals, k, None, criteria, 10, cv2.KMEANS_RANDOM_CENTERS) 
    
    # Os dados são convertidos para valores inteiros de 8 bits
    centers = np.uint8(centers) 
    segmented_data = centers[labels.flatten()]
    
    # Remodelando os dados para o tamanho da imagem original
    segmented_image = segmented_data.reshape((image.shape)) 
    
    return segmented_image

# Função para plotar imagens com subplots
def image_show(image, nrows=1, ncols=1, cmap='gray'):
    fig, ax = plt.subplots(nrows=nrows, ncols=ncols, figsize=(14, 14))
    ax.imshow(image, cmap='gray')
    ax.axis('off')
    return fig, ax


def circle_points(resolution, center, radius): 
    
    # Gera um conjunto de pontos que representa um círculo sobre a imagem 
    radians = np.linspace(0, 2*np.pi, resolution)
    c = center[1] + radius*np.cos(radians)#polar co-ordinates
    r = center[0] + radius*np.sin(radians)
    
    return np.array([c, r]).T

def luminosity(img):
    img_gray = (0.2126 * img[:, :, 0] + 0.7152 * img[:, :, 1] + 0.0722 * img[:, :, 2]).astype(np.uint8)
    return img_gray
```

Se tentássemos fazer a segmentação utilizando theshold supervisionado:

```python
threshold = 10
(l,c) = bible_gray.shape
abs_tg_c = np.abs(np.diff(bible_gray.astype(np.float64), axis=0, append=255))
abs_tg_l = np.abs(np.diff(bible_gray.astype(np.float64), axis=1, append=255))
abs_tg = np.sqrt(abs_tg_c ** 2 + abs_tg_l ** 2)
ret, threshold = cv2.threshold(bible_gray,0,255,cv2.THRESH_BINARY_INV+cv2.THRESH_OTSU)
bible_segment = np.zeros(shape=(l, c), dtype=np.uint8)
bible_segment[abs_tg > threshold] = 255

plt.figure(figsize=(8, 8))
plt.imshow(bible_segment,cmap = 'gray')
```

Como pode ser observado, a qualidade não fica boa, uma vez que é difícil escolher o valor adequado de threshold.

O Scikit-image possui diversos métodos de thresholding automáticos, que não necessitam que algum valor seja fornecido. Alguns desses métodos são: otsu, li e local.
Para o exemplo, utilizaremos o local, sendo também necessário especificar o block_size. O offset ajuda a melhorar os resultados obtidos.

```python

bible_gray = luminosity(bible_2)
bible_threshold = filters.threshold_local(bible_gray,block_size=51, offset=10) 

plt.figure(figsize=(8, 8))
plt.imshow(bible_gray > bible_threshold, cmap = 'gray')
```

```python
# Convertendo o espaço de cores para HSV
hsv_nemo = cv2.cvtColor(nemo, cv2.COLOR_RGB2HSV)

# Definindo os limites de cores que desejam ser apresentados
light_orange = (1, 190, 200)
dark_orange = (50, 255, 255)

# Criando a máscara a partir dos limites de cores
mask = cv2.inRange(hsv_nemo, light_orange, dark_orange)

# Aplicando a máscara utilizando a função AND (E) bit a bit
result = cv2.bitwise_and(nemo, nemo, mask=mask)

plt.figure(figsize=(8, 8))
plt.imshow(result)
```

Segmentação de Contorno Ativo (Snakes):

Este tipo de segmentação é inicializado usando um cortorno ou linha definido pelo usuário sobre a área de interesse. Esse contorno então se contrai lentamente e é atraído ou repelido por luzes e arestas.

Para o exemplo, iremos realizar a segmentação do rosto de uma pessoa.

O algoritmo calcula as coordenas x e y dos pontos ao longo da circunferência. Como foi passada uma resolção de 200, serão plotados 200 pontos para formar o círculo.

Assim, o rosto é segmentado do resto da imagem a partir do ajuste de uma curva (snake) com os contornos do rosto. 

```python
points = circle_points(200, [290, 670], 270)[:-1]

fig, ax = image_show(bean)
ax.plot(points[:, 0], points[:, 1], '--r', lw=3)
```

```python
gray = color.rgb2gray(bean)
snake = seg.active_contour(gray, points,alpha=0.06,beta=0.1)
fig, ax = image_show(bean)
ax.plot(points[:, 0], points[:, 1], '--r', lw=3)
ax.plot(snake[:, 0], snake[:, 1], '-b', lw=3);
```
